```python
import duckdb

con = duckdb.connect("file.db")

con.sql("CREATE TABLE IF NOT EXISTS employees AS SELECT * FROM read_csv_auto('employees.csv')")
con.sql("CREATE TABLE IF NOT EXISTS users (username VARCHAR, age INTEGER, country VARCHAR)")

con.table("employees").show()
con.table("users").show()

con.close()
```

```python
import requests

users = [
    {"username": "alice_nyc",  "age": 28, "country": "USA"},
    {"username": "bob_london", "age": 34, "country": "UK"},
    {"username": "carlos_mx",  "age": 22, "country": "Mexico"},
    {"username": "diana_fr",   "age": 45, "country": "France"},
    {"username": "evan_us",    "age": 30, "country": "USA"},
    {"username": "fiona_ca",   "age": 27, "country": "Canada"},
    {"username": "george_uk",  "age": 38, "country": "UK"},
]

for user in users:
    r = requests.post("http://127.0.0.1:5000/users", json=user)
    print(r.status_code, r.json())
```

```python
r = requests.get("http://127.0.0.1:5000/users/summary")
print(r.status_code)
print(r.json())
```
